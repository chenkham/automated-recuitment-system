# Feature Engineering & Scoring Dry Run

This notebook demonstrates how candidate features (role fit, experience fit, company tier, and behavioral signals) are loaded and scored using our modular pipeline.

```python
import sys
sys.path.append('..')  # Add parent directory to path

from src.data_loader import stream_candidates
from src.filters import should_filter_out
from src.scorer import CandidateScorer
```

## Initialize Scorer and Load Sample Candidate

We'll load a small subset of the raw candidates, check if they pass hard filters, and compute their individual feature scores.

```python
CANDIDATES_PATH = '../data/raw/[PUB] India_runs_data_and_ai_challenge/India_runs_data_and_ai_challenge/candidates.jsonl'
scorer = CandidateScorer()
jd_text = "Senior AI Engineer with embeddings, retrieval, and vector search experience at product startups."

sample_count = 0
for c in stream_candidates(CANDIDATES_PATH):
    if should_filter_out(c):
        continue
        
    # Compute component scores
    role_fit = scorer.compute_role_fit(c, jd_text)
    exp_fit = scorer.compute_experience_fit(c)
    comp_fit = scorer.compute_company_fit(c)
    loc_fit = scorer.compute_location_fit(c)
    final_score = scorer.score_candidate(c, jd_text)
    
    print(f"=== Candidate: {c['candidate_id']} ===")
    print(f"Current Title: {c['profile']['current_title']}")
    print(f"Years of Experience: {c['profile']['years_of_experience']}")
    print(f"Role Fit Score: {role_fit:.2f}")
    print(f"Experience Fit Score: {exp_fit:.2f}")
    print(f"Company Fit Score: {comp_fit:.2f}")
    print(f"Location Fit Score: {loc_fit:.2f}")
    print(f"Final Score (with multipliers): {final_score:.2f}")
    print("\n")
    
    sample_count += 1
    if sample_count >= 3:
        break
```
